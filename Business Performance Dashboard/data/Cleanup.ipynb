```python
path = r'C:\Users\aekal\PycharmProjects\powerbi\Data'
file = 'Online_Sales_fin.csv'
full_path = os.path.join(path, file)
```

```python
df = pd.read_csv(full_path, parse_dates=True)
```

```python
df['Avg_Price']=df['Avg_Price'].astype('int')
df['Delivery_Charges']=df['Delivery_Charges'].astype('int')
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
df = df.set_index('Transaction_Date')
```

```python
df['Month'] = df.index.month
```

```python
df['Month_Name'] = df.index.strftime('%b')
```

```python
df.reset_index()
```

```python
df.to_csv('Online_Sales_Clean.csv')
```
